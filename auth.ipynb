```python
def opensite(url):
    # OPEN SITE

    driver = webdriver.Chrome(r"C:\Users\Giorgos Magginas\chromedriver\chromedriver.exe")
    driver.get(url)

    sleep(random.uniform(2, 3))
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Ok')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Accept')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
        
    
    return driver
```
