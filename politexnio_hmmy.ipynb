```python
import time
from time import sleep
from tqdm import tqdm


#---------Selenium------------

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# generate random for time
import random

# for dates
from datetime import datetime, timedelta

# regex of course
import re

# pandas for dataframe
import pandas as pd

# to create .pkl
import joblib

# Import date class from datetime module and get  the current date
from datetime import date

# for create  a new directory
import os
```

```python
def opensite(url):
    # OPEN SITE

    driver = webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")
    driver.get(url)

    sleep(random.uniform(2, 3))
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Ok')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Accept')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
        
    
    return driver
```

```python
url="https://www.ece.ntua.gr/gr/announcements"
driver=opensite(url)
```

```python
all_announcements=driver.find_element_by_tag_name('tbody').find_elements_by_xpath("//tr[@class='clickable-row']")
```

```python
all_announcements[0].text.split('\n')
```

```python
# create a df
cols = ['date', 'title', 'category', 'title_of_announcement', 'thetext']

df = pd.DataFrame(columns=cols)
```

```python
for i in range(len(all_announcements)):
    print(i)
    driver.get(url)
    all_announcements=driver.find_element_by_tag_name('tbody').find_elements_by_xpath("//tr[@class='clickable-row']")
    announcements = all_announcements[i].text.split('\n')
#     print(announcements)
    if  announcements[2]=='Προκηρύξεις':
        date = announcements[0]
        title = announcements[1]
        category = announcements[2]
        
        sleep(random.uniform(1, 2))
        
        all_announcements[i].click()
        title_of_announcement=driver.find_element_by_xpath("//div[@class='text']").find_element_by_tag_name('h3').text
        text=""
        paragraphs=driver.find_element_by_xpath("//div[@class='text']").find_elements_by_tag_name('p')
        for i in range(len(paragraphs)):
            if text=="":
                thetext=paragraphs[i].text
            else:
                thetext=thetext+" \n "+paragraphs[i].text
                            # complete dataset
        temp = {
            'date' : date,
            'title' : title,
            'category' : category,
            'title_of_announcement' : title_of_announcement,
            'thetext' : thetext
        }

        df = df.append(temp, ignore_index=True,)
        sleep(random.uniform(1, 2))
        driver.get(url)
```
