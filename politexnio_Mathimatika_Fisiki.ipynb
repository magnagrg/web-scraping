```python
import time
from time import sleep
from tqdm import tqdm


#---------Selenium------------

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# generate random for time
import random

# for dates
from datetime import datetime, timedelta

# regex of course
import re

# pandas for dataframe
import pandas as pd

# to create .pkl
import joblib

# Import date class from datetime module and get  the current date
from datetime import date

# for create  a new directory
import os
```

```python
def opensite(url):
    # OPEN SITE

    driver = webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")
    driver.get(url)

    sleep(random.uniform(2, 3))
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Ok')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Accept')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
        
    
    return driver
```

```python
url = 'http://semfe.ntua.gr/el/'
driver=opensite(url)
```

```python
# Find  the tap line  
tapsline=driver.find_element_by_xpath("/html/body/div[1]/div[2]/div/div/div/div/section/div/div/div/div/ol")
category = tapsline.find_elements_by_tag_name('li')
# and take tap prokiriksis
for i in range(len(category)):
    if category[i].get_attribute("id")==  'prokirixeis':   
        tapforclick=category[i]
        
sleep(random.uniform(1, 2))
tapforclick.click()
sleep(random.uniform(2, 3))
# # katw sto telos  for  more

footer = driver.find_element_by_xpath("//a[@title='Περισσότερα από Προκηρύξεις Θέσεων']")
thelink = footer.get_attribute("href")

sleep(random.uniform(1, 2))
driver.get(thelink)

# the date
dates=driver.find_elements_by_xpath("//div[@style='line-height:1.2;margin-top:2px;']")

# the content
content=driver.find_elements_by_xpath("//a[@style='margin-top:2px !important;']")

# for the link of cntent
# content[1].get_attribute("href")

# find the number of the pages 
how_many_pages =  driver.find_element_by_xpath("//li[@class='counter']")
how_many_pages=int(how_many_pages.text.split()[-1])

# the next button
next_page= driver.find_element_by_xpath("//a[@title='Επόμενο']")
```

```python
# create a df
cols = ['dates', 'content', 'Link', 'pdfLink']

df = pd.DataFrame(columns=cols)
```

```python
for i  in range(1,how_many_pages+1):
    print('page '+str(i))


    
    for j in range(len(dates)):
        print('prokiriksi {}'.format(j))
        thedate = dates[j].text
        thecontent = content[j].text
        thelink = content[j].get_attribute("href")
        driver2=opensite(thelink)
        try:
            thepdf= driver2.find_element_by_xpath("//ul[@class='itemAttachments']")
        except:
            try:
                thepdf=driver2.find_element_by_xpath("//a[@class='itemPrevious']").click()
                thepdf= driver2.find_element_by_xpath("//ul[@class='itemAttachments']")
            except:
                thepdf= driver2.find_element_by_xpath("//div[@class='itemFullText']")
            
        pdflink = thepdf.find_element_by_tag_name('a').get_attribute("href")
        driver2.quit()
        temp = {
            'dates':thedate,
            'content':thecontent,
            'Link':thelink,
            'pdfLink' : pdflink
        }
        df = df.append(temp, ignore_index=True,) 
    if i == how_many_pages:
        driver.quit()
        break
    elif i == how_many_pages-1:
        sleep(random.uniform(1, 2))
        next_page.click()
        # the date
        dates=driver.find_elements_by_xpath("//div[@style='line-height:1.2;margin-top:2px;']")

        # the content
        content=driver.find_elements_by_xpath("//a[@style='margin-top:2px !important;']")

    else:
        sleep(random.uniform(1, 2))
        next_page.click()
        # the date
        dates=driver.find_elements_by_xpath("//div[@style='line-height:1.2;margin-top:2px;']")

        # the content
        content=driver.find_elements_by_xpath("//a[@style='margin-top:2px !important;']")

        # the next button
        next_page= driver.find_element_by_xpath("//a[@title='Επόμενο']")
```
