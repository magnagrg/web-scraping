```python
import time
from time import sleep
from tqdm import tqdm


#---------Selenium------------

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# generate random for time
import random

# for dates
from datetime import datetime, timedelta

# regex of course
import re

# pandas for dataframe
import pandas as pd

# to create .pkl
import joblib

# Import date class from datetime module and get  the current date
from datetime import date

# for create  a new directory
import os
```

```python
def opensite(url):
    # OPEN SITE

    driver = webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")
    driver.get(url)

    sleep(random.uniform(2, 3))
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Ok')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Accept')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
        
        
    
    return driver
```

```python
# # the main site
# url=f'https://www.unic.ac.cy/'
# # for masters
# urlmasters = 'https://www.unic.ac.cy/master-degrees/'
# # for doctorates
# urldoctorates = 'https://www.unic.ac.cy/doctoral-degrees/'
```

```python
# create a df
cols = ['DEPARTMENT', 'PROGRAMME', 'linkPROGRAMME', 'LANGUAGE OF INSTRUCTION', 'DEGREE','Duration', 'Qualification Awarded', 'Level of Qualification', 'Mode of Study', 'Minimum ECTS Credits']

df = pd.DataFrame(columns=cols)
```

```python
# for masters
urlmasters = 'https://www.unic.ac.cy/master-degrees/'
# for doctorates
urldoctorates = 'https://www.unic.ac.cy/doctoral-degrees/'
# masters_and_doctorates
masters_and_doctorates=['https://www.unic.ac.cy/master-degrees/','https://www.unic.ac.cy/doctoral-degrees/']

# elegxoi  gia selides
flag=True
flag1=True

for  thedegree in masters_and_doctorates:
    driver=opensite(thedegree)
    tapsline=driver.find_element_by_xpath("/html/body/div[1]/div[2]/main/div/section/div/div/div[2]/div/div/div/div[1]/div[1]/ul")
    taps=tapsline.find_elements_by_tag_name("li")
    if  thedegree==masters_and_doctorates[0]:
        DEGREE='Master'
        
    elif  thedegree==masters_and_doctorates[1]:
        DEGREE='Doctoral'
    

    for numdepartment in range(len(taps)):
        sleep(random.uniform(1,1.5))
        taps[numdepartment].click()
#         print(numdepartment)
#         print(taps[numdepartment].text)

        # DEPARTMENT
        DEPARTMENT=taps[numdepartment].text



        small_table=driver.find_element_by_xpath(f'/html/body/div[1]/div[2]/main/div/section/div/div/div[2]/div/div/div/div[1]/div[2]/div[{((numdepartment+1)*2)}]')
        courses=small_table.find_elements_by_tag_name("article")
        for i in range(len(courses)):

            # PROGRAMME
            PROGRAMME=courses[i].text

            # linkPROGRAMMElist
            linkPROGRAMMElist=courses[i].find_element_by_tag_name('a').get_attribute("href")

            
            driver2=opensite(linkPROGRAMMElist)
            try:
                minitable=driver2.find_element_by_xpath('/html/body/div[1]/div[2]/main/div/section/article/div/div[3]/div/div[1]/div/div[2]/table')
            except:
                try:
                    minitable=driver2.find_element_by_xpath('/html/body/div[1]/div[2]/main/div/section/article/div/div[2]/div/div[1]/div/div[2]/table')
                except:
                    try:
                        minitable=driver2.find_element_by_xpath('/html/body/div[1]/div[2]/main/div/section/div/div/div[2]/div/div/div/div[5]')
                        minirows=minitable.find_elements_by_tag_name('h4')                    
                        flag=False
                    except:
                        flag1=False
                        
            if flag1:
                if flag:

                    minirows=minitable.find_elements_by_tag_name('tr')


                    # LANGUAGE
                    LANGUAGE=minirows[3].find_elements_by_tag_name('td')[1].text

                    # Duration
                    Duration=minirows[0].find_elements_by_tag_name('td')[1].text

                    # Qualification
                    Qualification=minirows[1].find_elements_by_tag_name('td')[1].text

                    # LevelofQualification
                    LevelofQualification=minirows[2].find_elements_by_tag_name('td')[1].text

                    # ModeofStudy
                    ModeofStudy = minirows[4].find_elements_by_tag_name('td')[1].text

                    # MinimumECTSCredits
                    MinimumECTSCredits = minirows[5].find_elements_by_tag_name('td')[1].text

                else:

                    flag=True

                    # LANGUAGE
                    LANGUAGE=minirows[2].find_elements_by_tag_name('span')[1].text

                    # Duration
                    Duration=minirows[0].find_elements_by_tag_name('span')[1].text

                    # Qualification
                    Qualification=minirows[1].find_elements_by_tag_name('span')[1].text

                    # LevelofQualification
                    LevelofQualification=DEGREE

                    # ModeofStudy
                    ModeofStudy = minirows[3].find_elements_by_tag_name('span')[1].text

                    # MinimumECTSCredits
                    MinimumECTSCredits = ''
            else:
                flag1=True
                
                # LANGUAGE
                LANGUAGE=""
                
                # Duration
                Duration=""
                
                # Qualification
                Qualification=""
                
                # LevelofQualification
                LevelofQualification=DEGREE

                # ModeofStudy
                ModeofStudy = ""
                
                # MinimumECTSCredits
                MinimumECTSCredits = ''
                

            driver2.quit()

            temp = {
                'DEPARTMENT':DEPARTMENT,
                'PROGRAMME':PROGRAMME,
                'linkPROGRAMME':linkPROGRAMMElist,
                'LANGUAGE OF INSTRUCTION':LANGUAGE,
                'DEGREE':DEGREE,
                'Duration':Duration, 
                'Qualification Awarded':Qualification, 
                'Level of Qualification':LevelofQualification, 
                'Mode of Study':ModeofStudy, 
                'Minimum ECTS Credits':MinimumECTSCredits
            }
            df = df.append(temp, ignore_index=True,)

```
