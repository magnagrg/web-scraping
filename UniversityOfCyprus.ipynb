```python
def opensite(url):
    # OPEN SITE

    driver = webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")
    driver.get(url)

    sleep(random.uniform(2, 3))
    try:
        driver.find_element_by_xpath("//*[contains(text(), 'Έγινε')]").click()

    except:
        pass
        sleep(random.uniform(0,0.5))
    
    return driver
```

```python
url='https://www.ucy.ac.cy/graduateschool/index.php/el/postgraduate-stuties/postgraduate-programmes-of-study#Gen.Info'
url='https://www.ucy.ac.cy/graduateschool/en/postgraduate-studies/postgraduate-programmes-of-study'
driver=opensite(url)
```

```python
# create a df
cols = ['DEPARTMENT','linkDEPARTMENT' , 'PROGRAMME', 'linkPROGRAMME', 'LANGUAGE OF INSTRUCTION', 'DEGREE']

df = pd.DataFrame(columns=cols)
```

```python
# tha perimeni  gia 350 seconds mexri na piasei tin timi
# driver.implicitly_wait(350) 

table=driver.find_element_by_xpath("/html/body/section[3]/section/div[1]/div[1]/div[2]/div[3]/table/tbody")
rows=table.find_elements_by_tag_name('tr')
rows=rows[1:]
for i in range(len(rows)):
    cell=rows[i].find_elements_by_tag_name('td')
    
    if len(cell)==4:       
        DEPARTMENT=cell[0].text
        
#         --------------------------------------------------------
        linkDEPARTMENT= cell[0].find_elements_by_tag_name('p')
        
        if len(linkDEPARTMENT)== 0 or len(linkDEPARTMENT)==1:
            try:
                linkDEPARTMENT= cell[0].find_element_by_tag_name('a').get_attribute("href")
            except:
                linkDEPARTMENT="no link available"
        else:
            linkDEPARTMENTlist=[]        
            for i in range(len(linkDEPARTMENT)):
                try:
                    linkDEPARTMENTlist.append(linkDEPARTMENT[i].find_element_by_tag_name('a').get_attribute("href"))
                except:
                    linkDEPARTMENTlist.append("no link available")       
                    
#         --------------------------------------------------------
        PROGRAMME=cell[0].text.split('\n')
        if len(PROGRAMME)==1:
            PROGRAMME=PROGRAMME[0]
    
#         --------------------------------------------------------
        linkPROGRAMME= cell[1].find_elements_by_tag_name('p')
        
        if len(linkPROGRAMME)== 0 or len(linkPROGRAMME)==1:
            try:
                linkPROGRAMMElist= cell[0].find_element_by_tag_name('a').get_attribute("href")
            except:
                linkPROGRAMMElist="no link available"
        else:
            linkPROGRAMMElist=[]        
            for i in range(len(linkPROGRAMME)):
                try:
                    linkPROGRAMMElist.append(linkPROGRAMME[i].find_element_by_tag_name('a').get_attribute("href"))
                except:
                    linkPROGRAMMElist.append("no link available")

#         --------------------------------------------------------
        LANGUAGE=cell[2].text.split('/')
        if len(LANGUAGE)==1:
            LANGUAGE=LANGUAGE[0]
            
#         --------------------------------------------------------
        DEGREE=cell[3].text.split('/')
        if len(DEGREE)==1:
            DEGREE=DEGREE[0]

#         --------------------------------------------------------
    else:
        PROGRAMME=cell[0].text.split('\n')
        if len(PROGRAMME)==1:
            PROGRAMME=PROGRAMME[0]
        
#         --------------------------------------------------------
        linkPROGRAMME= cell[0].find_elements_by_tag_name('p')
        
        if len(linkPROGRAMME)== 0 or len(linkPROGRAMME)==1:
            try:
                linkPROGRAMMElist= cell[0].find_element_by_tag_name('a').get_attribute("href")
            except:
                linkPROGRAMMElist="no link available"
        else:
            linkPROGRAMMElist=[]        
            for i in range(len(linkPROGRAMME)):
                try:
                    linkPROGRAMMElist.append(linkPROGRAMME[i].find_element_by_tag_name('a').get_attribute("href"))
                except:
                    linkPROGRAMMElist="no link available"

            
#         --------------------------------------------------------
        LANGUAGE=cell[1].text.split('/')
        if len(LANGUAGE)==1:
            LANGUAGE=LANGUAGE[0]
    
#         --------------------------------------------------------
        DEGREE=cell[2].text.split('/')
        if len(DEGREE)==1:
            DEGREE=DEGREE[0]
    
#         --------------------------------------------------------
        
#     load to dataset
    temp = {
        'DEPARTMENT':DEPARTMENT,
        'linkDEPARTMENT':linkDEPARTMENT,
        'PROGRAMME':PROGRAMME,
        'linkPROGRAMME':linkPROGRAMMElist,
        'LANGUAGE OF INSTRUCTION':LANGUAGE,
        'DEGREE':DEGREE
    }
    df = df.append(temp, ignore_index=True,)
# driver.quit()
```
